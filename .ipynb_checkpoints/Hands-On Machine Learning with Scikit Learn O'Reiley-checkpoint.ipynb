# Linear Regression

Normal Equation : $\hat\theta = (X^T.X)^{-1}.X^T.y$

$\hat\theta$ is the best $\theta$ that minimizes the cost function

```python
import numpy as np
import pandas as pd 
import random
import matplotlib.pyplot as plt
```

```python
X = 2*np.random.rand(100,1)
y = 4 + 3*X + np.random.rand(100,1)
```

```python
plt.scatter(X,y,marker = '.')
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
```

```python
plt.plot(X_new,y_predict, 'r')
plt.scatter(X,y,marker = '.')
plt.show()
```

# Gradient Descent

$\Delta_\theta(MSE(\theta)) = \frac{2}{m}X^T.(X.\theta - y)$

$\theta^{ next step} = \theta - \eta \Delta_\theta(MSE(\theta))$

```python
eta = 0.1
n_iterations = 1000
m = 100
theta = np.random.rand(2,1)
plt.scatter(X,y,marker = '.')
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta)-y)
    theta = theta - eta*gradients  
    y_predict = X_new_b.dot(theta)
    plt.plot(y_predict,X_new)
    
```

# Stochastic Gradient Descent

SGD just picks a random instance in the training set at every step and computes the gradients based only on that single instance.

```python
n_epoch = 50
t0 = 5
t1 = 50
m = 100

def learningRateChanger(t):
    return t0/(t+t1)

theta = np.random.rand(2,1) #random initialization of theta vector

for epoch in range(n_epoch):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learningRateChanger(epoch*m + i)
        theta = theta - eta*gradients
```

```python
theta
```

# Stochastic Gradient Descent using Sklearn

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter=50, penalty = None, eta0 = 0.1)
sgd_reg.fit(X,y.ravel())
```
